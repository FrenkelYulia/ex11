```python
import pandas as pd
```

```python
#Задание 1
'''Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. 
Определите какому фильму было выставлено больше всего оценок 5.0.'''

ratings = pd.read_csv('ratings.csv')
ratings.head()
r= ratings.loc[ratings.rating==5.0].groupby('movieId').count().sort_values('rating', ascending=False).head(1)
r = r.reset_index()
print(r)
number = r['movieId']
print('было выставлено больше всего оценок 5.0  фльму с ID:', sum(number))

```

```python
#Задание 2
'''По данным файла power.csv посчитайте суммарное потребление стран Прибалтики
(Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года.
Не учитывайте в расчетах отрицательные значения quantity.'''
data = pd.read_csv('power.csv')
data.head()
#data[ data['country'].str.contains('es', case=False) ]['country'].unique()
filter1 = data['category'].isin([4,12,21])
filter2 = data['country'].isin(['Lithuania','Latvia','Estonia'])
filter3 = (data['year']>2005)&(data['year']<2010)
filter4 = data['quantity']>0
new_data = data.loc[filter1&filter2&filter4&filter3]
new_data.head()
sum_quant = new_data['quantity'].sum()
print(sum_quant)  
```
